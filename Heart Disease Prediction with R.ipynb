```python
### In R programming, the mutate function is used to create a new variable from a data set. 
##  In order to use the function, we need to install the dplyr package, which is an add-on to R that includes a host of cool functions for selecting, filtering, grouping, and arranging data
```

```python
data2 <- data %>%
mutate( sex = if_else (sex == 1, "MALE", "FEMALE"),
       fbs = if_else (fbs == 1, ">120", "<=120" ),
       exang = if_else (fbs == 1, "YES", "NO"),
       cp = if_else (cp == 1, "ATYPICAL ANGINA",
                   if_else(cp == 2, "NON-ANGINAL PAIN", "ASYMPTOMATIC")),
       restecg = if_else( restecg==0, "NORMAL",
                    if_else(restecg==1, "ABNORMALITY", "PROBABLE OR DEFINITE")),
       slope = as.factor(slope),
       ca = as.factor(ca),
       thal = as.factor(thal),
       target = if_else(target ==1, "YES", "NO") ## if the target is 1 then you have a heart disease else you dont have.
       ) %>%
mutate_if(is.character, as.factor)%>% ## here you change 
dplyr::select(target, sex, fbs, exang,cp, restecg, slope, ca, thal, everything()) ### you do data manipulation using the variables in the bracket.

### after running the above code, it will note produce the output here rather it has already been processed by R internally.
### I need to fully understand the above code 
```

```python
 #### Data Visualization
### we shall use ggplot 

## Bar plot for target (Heart Disease)


### aes means Construct Aesthetic Mappings.Aesthetic mappings describe how variables in the data are mapped to visual properties (aesthetics) of geoms.
 ggplot(data2, aes(x=data2$target, fill=data2$target)) +
        geom_bar()+
        xlab("Heart Disease")+
        ylab("Count")+
ggtitle("Presence and Absence of Heart Disease")
scale_fill_discrete(name="Heart Disease", lables=c("Absence", "Presence"))
```
